```python
keys = list(keys)
keys[0]
```

```python
content[keys[0]]['noun']
```

```python
def single_detect(single_slice):
    for i in range(2):
        for noun in content[keys[i]]['noun']:
            if noun not in single_slice:
                print('not in %s-noun' % keys[i])
                break
            else:
                # 进入下一层: 看是否不符合相关自行车警情
                for ban in content[keys[i]]['overcome']:
                    if ban in single_slice:
                        print('not in %s-overcome ' % keys[i])
                        break
                    else:
                        # 判断该条是否确切在含有动词，或者词组如果有就是，没有就是其他。
                        verb_grp = content[keys[i]]['verb']
                        for verb in verb_grp.extend(content[keys[i]]['group']):
                            if verb in single_slice:
                                print('in %s' % keys[i])
                            else:
                                print('not in %s-verb&verbgroup' % keys[i])
                            
```
